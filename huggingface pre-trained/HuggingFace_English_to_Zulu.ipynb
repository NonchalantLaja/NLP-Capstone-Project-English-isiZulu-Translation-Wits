```python
# Importing the pipeline function from the transformers library
from transformers import pipeline
```

```python
# Creating a Text2TextGenerationPipeline for language translation
pipe = pipeline(task='text2text-generation', model='facebook/m2m100_418M')
```

```python
# Converting 
pipe("You are Beautiful", forced_bos_token_id=pipe.tokenizer.get_lang_id(lang='zu'))
```

```python
def translate(text):
  return(pipe(text, forced_bos_token_id=pipe.tokenizer.get_lang_id(lang='zu')))
```

```python
def turn_to_str(val):
  return "".join(val)
```

```python
from nltk.translate.bleu_score import sentence_bleu
#from nltk.translate.bleu_score import SmoothingFunction
#smooth = SmoothingFunction().method2 # math domain error with method4

def calculate_bleu(x):
    hypothesis = x['hypothesis'].split()
    reference = x['reference'].split()

    reference_list = [reference] # list of references for 1 sentence.
    bleu_score = sentence_bleu(reference_list, hypothesis)#,smoothing_function = smooth)
    return bleu_score
```

```python
sub['score'] = sub[['hypothesis', 'reference']].apply(calculate_bleu, axis=1)
```

```python
sub.head()
```

```python
print("Average Bleu Score for 50 samples: {}".format(sum(sub['score'] / len(sub['score']))) )
```

## Using pre-trained mBart50 model

```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
```

```python
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-one-to-many-mmt")
```

```python
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-one-to-many-mmt", src_lang="en_XX")
```

```python
article_en = "U.N encourages wearing masks"
```

```python
model_inputs = tokenizer(article_en, return_tensors="pt")
```

```python
# translate from English to Zulu
generated_tokens = model.generate(
    **model_inputs,
    forced_bos_token_id=tokenizer.lang_code_to_id["af_ZA"] ## language code for Zulu ??
)
```

```python
translation = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```
